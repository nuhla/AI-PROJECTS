```python
nlp= spacy.load("en_core_web_sm")

doc = nlp("We just Opend our ings, the flying part is comming soon")

for token in doc:
    if token.is_stop:
        print(token.text)
```

```python
def preprocess(text):
    doc = nlp(text)

    no_stop =[token.text for token in doc if not token.is_stop and not token.is_punct]
    return " ".join(no_stop)
 
```

```python
preprocess("We just Opend our ings, the flying part is comming soon")
```
